#### Question 7 (1 point)
What are the different types of loops, and when might you want to try one over another? Script an example.

=== BEGIN MARK SCHEME ===  
1. __For loop__

```ruby
%%bash
# For / While / Util

for i in {1..3}
do
  echo ${i}
done
```
```
1
2
3
```

2. __While loop__

```ruby
%%bash
i=0

while [ ${i} -lt 3 ]
do
  echo ${i}
  i=$((i+1))
done
```
```
0
1
2
```

3. __Until loop__  

```ruby
%%bash
i=0

until [ ${i} -gt 3 ]
do
  echo ${i}
  i=$((i+1))
done
```
```
0
1
2
3
```
